```python
#1

import random,time
def sort_simple_selection(seq):
    for i in range(len(seq)-1):
        min = i
        for j in range(i+1,len(seq)):
            if seq[j] < seq[min]:
                min = j
        seq[i],seq[min] = seq[min],seq[i]
#
num = []
start_time = time.time()
for i in range(10000):
    num.append(random.randrange(100000))
sort_simple_selection(num)
print(num)
end_time = time.time()
print(end_time - start_time)
```

```python
import random,time


def quick_sort(seq):
    left_seq = []
    right_seq = []
    p = seq[0]
    for number in seq[1:]:
        if number <= p:
            left_seq.append(number)
        else:
            right_seq.append(number)
    if left_seq:
        left_seq = quick_sort(left_seq)
    if right_seq:
        right_seq = quick_sort(right_seq)
    return left_seq +[p] +right_seq

num = []
start_time = time.time()
for i in range(10000):
    num.append(random.randrange(100000))
a = quick_sort(num)
print(a)
end_time = time.time()
print(end_time - start_time)
```
