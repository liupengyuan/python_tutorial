```python
def jiecheng(n):
    i=1
    p=1
    while i < n+1:
        p=p*i
        i=i+1    
    return p
m = int(input('请输入第1个整数 '))
n = int(input('请输入第2个整数 '))
k = int(input('请输入第3个整数 '))
print('结果为：',jiecheng(m)+jiecheng(n)+jiecheng(k))
```

```python
def he(n):
    s=0
    i=1
    while i<n+1:
        if (i%2==0):
            s=s-1/(2*i-1)
        else:
            s=s+1/(2*i-1)
        i=i+1
    return s

n=1000
he1=4*he(n)
print(he1)

n=10000
he2=4*he(n)
print(he2)
```
