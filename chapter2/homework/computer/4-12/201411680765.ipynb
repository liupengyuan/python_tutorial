```python

#写函数，返回一个list中所有数字的和

    
def create_list():
    n = int(input('请输入您要键入的数字个数：'))
    numbers = []
    i = 0
    while i < n:
        number = int(input('请输入一个数字：'))
        numbers.append(number)
        i += 1
    return numbers

def sum_of_list(numbers):
    total = 0
    for number in numbers:
        total = total + number
    return total
    #return sum(numbers)



print (sum_of_list(create_list()))

    
```

```python
#写函数，返回一个list中的最小值


def create_list():
    n = int(input('请输入您要键入的数字个数：'))
    numbers = []
    i = 0
    while i < n:
        number = int(input('请输入一个数字：'))
        numbers.append(number)
        i += 1
    return numbers

def min_of_list(numbers):
    min_num = numbers[0]
    for number in numbers:
        if min_num > number:
            min_num = number
    return min_num
    #return min(numbers)

print(min_of_list(create_list()))
```

```python
#写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.

def location(n,numbers):
    i = 0
    for number in numbers:
        if number == n:
            return i
        else:
            i += 1
            
numbers = [5,6,7,1,2,3,4]
print (location(3,numbers))
```

```python
#写函数，可求两个向量的夹角余弦值，向量可放在list中。主程序调用该函数
import math

def cos_of_angle(xiangliang1,xiangliang2):
    multi_of_mo = (math.sqrt(xiangliang1[0]**2+xiangliang1[1]**2))*(math.sqrt(xiangliang2[0]**2+xiangliang2[1]**2))
    cos = (xiangliang1[0]*xiangliang2[0]+xiangliang1[1]*xiangliang2[1])/multi_of_mo
    return cos

xiangliang1 = [0,1]
xiangliang2 = [0,1]    
print(cos_of_angle(xiangliang1,xiangliang2))

```

```python
#挑战性习题：python语言老师为了激励学生学python，自费买了100个完全相同的Macbook Pro，分给三个班级，每个班级至少分5个，
#用穷举法计算共有多少种分法？

def methods_of_distribution(n):
    methods = 0
    for i in range(5,n+1-5-5):
        for j in range(5,n+1-5-5):
            for k in range(5,n+1-5-5):
                if i + j + k == 100:
                    print ('三个班分别分得：','i=',i,'j=',j,'k=',k)
                    methods += 1     
    
    return methods

methods = methods_of_distribution(100)
print ('共有',methods,'种方法')

```
