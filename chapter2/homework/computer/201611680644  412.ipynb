```python
def compute_sum(end):
    i = 1
    total_n = 1

    for i in range(end):
        i = i + 1
        total_n = total_n * i

    return total_n

n = int(input('请输入第1个整数，以回车结束。'))
m = int(input('请输入第2个整数，以回车结束。'))
k = int(input('请输入第3个整数，以回车结束。'))

print('最终的和是：', compute_sum(m) + compute_sum(n) + compute_sum(k))
```
